```python
import os
from spleeter.separator import Separator

ruta_canciones = "/content/drive/MyDrive/MIR/mir_datasets/Anouk"
ruta_salida = "/content/drive/MyDrive/MIR/mir_datasets/Anouk"
os.makedirs(ruta_salida, exist_ok=True)

# Separador con 2 stems
separator = Separator('spleeter:2stems')

for archivo in os.listdir(ruta_canciones):
    if archivo.endswith(".mp3"):
        ruta_archivo = os.path.join(ruta_canciones, archivo)
        salida_archivo = os.path.join(ruta_salida, archivo.replace('.mp3', ''))
        print(f"Procesando {archivo}...")

        # Separar la canción
        separator.separate_to_file(ruta_archivo, salida_archivo)

print("Separación completada. Revisa la carpeta de resultados.")
```

```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt

def calcular_sincopa(archivo_audio):
    # Cargar el archivo de audio
    y, sr = librosa.load(archivo_audio, sr=None)
    # Extraer el tempo y pulsos básicos
    tempo, beats = librosa.beat.beat_track(y=y, sr=sr, trim=False)
    print(f"Tempo detectado: {tempo} BPM")
    # Calcular el onset envelope (envolvente de ataques)
    onset_env = librosa.onset.onset_strength(y=y, sr=sr, aggregate=np.median)
    # Convertir los beats a tiempos en segundos
    beat_times = librosa.frames_to_time(beats, sr=sr)
    # Calcular la distancia de cada onset al beat más cercano
    onsets = librosa.onset.onset_detect(y=y, sr=sr, units='time')
    distancias = []
    for onset in onsets:
        distancias.append(min(abs(onset - beat_times)))
    # Calcular la "síncopa" como la desviación promedio de los onsets del ritmo
    sincopa_promedio = np.mean(distancias)

    print(f"Síncopa promedio: {sincopa_promedio:.4f} segundos")
    return sincopa_promedio

# Archivo de audio de prueba
archivo = "Waltz.mp3"
nivel_sincopa = calcular_sincopa(archivo)
print(f"Nivel de síncopa: {nivel_sincopa}")
```
